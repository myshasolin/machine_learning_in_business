*Верно. Если многократно повторять эксперимент, то 95% выборочных средних значений будут попадать в рассчитанный доверительный интервал.*

*Неверно. Мы можем быть на 95% уверены, что истинное среднее значение принадлежит рассчитанному доверительному интервалу, но это не означает, что оно точно находится в этом интервале.*

*Неверно. Среднее значение в генеральной совокупности может находиться как выше, так и ниже нижней границы 95% доверительного интервала.*

#### 4. Часто на практике нулевая гипотеза отклоняется, и различия считаются статистически достоверными, если p < 0,05. Однако часто в статистике используется более жесткий критерий достоверности различий, например, при условии, что p < 0,01. Например, если исследователь решил, что α = 0,05, то и нулевая гипотеза будет отклоняться при условии, что p < 0,05. 

*Нулевая гипотеза может быть отклонена при любом значении уровня значимости $\alpha$, включая 0,05 и 0,01. Однако чем ниже выбранный уровень значимости, тем более жестким становится критерий для отклонения нулевой гипотезы и тем выше пороговое значение статистики для ее отклонения.*
